```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

my_sentence = ["I love you","The war is coming","That's bad.","He hates you"]

pos_lexicon = ["good","love","laugh","fun"]
neg_lexicon = ["bad","crisis","war","hate"]

for sentence in my_sentence:
    posw = 0
    negw = 0
    totw = 0
    for token in word_tokenize(sentence):
        totw = totw + 1
        if stemmer.stem(token) in pos_lexicon:
            posw = posw + 1
        if stemmer.stem(token) in neg_lexicon:
            negw = negw + 1
    sentiment = (posw - negw) / totw
    print(sentence, sentiment)

```
